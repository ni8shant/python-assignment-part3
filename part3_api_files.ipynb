```python
# Task 1 - File Read & Write Basics

print("TASK 1 - FILE READ & WRITE BASICS")
print("="*70 + "\n")

# Part A - Write

print("=== PART A - WRITE ===\n")

# Step 1: Write five lines to the file using write mode
try:
    with open("python_notes.txt", "w", encoding="utf-8") as file:
        file.write("Topic 1: Variables store data. Python is dynamically typed.\n")
        file.write("Topic 2: Lists are ordered and mutable.\n")
        file.write("Topic 3: Dictionaries store key-value pairs.\n")
        file.write("Topic 4: Loops automate repetitive tasks.\n")
        file.write("Topic 5: Exception handling prevents crashes.\n")
    
    print("File written successfully.")
except Exception as e:
    print(f"Error writing file: {e}")

# Step 2: Append two more lines using append mode
try:
    with open("python_notes.txt", "a", encoding="utf-8") as file:
        file.write("Topic 6: Functions promote code reusability.\n")
        file.write("Topic 7: File I/O enables data persistence.\n")
    
    print("Lines appended successfully.")
except Exception as e:
    print(f"Error appending to file: {e}")

# Part B - Read

print("\n=== PART B - READ ===\n")

# Step 1: Read the file and print each line numbered, stripping newline
try:
    with open("python_notes.txt", "r", encoding="utf-8") as file:
        lines = file.readlines()
    
    print("File contents (numbered):\n")
    for i, line in enumerate(lines, 1):
        print(f"{i}. {line.strip()}")
    
except FileNotFoundError:
    print("Error: File 'python_notes.txt' not found.")
except Exception as e:
    print(f"Error reading file: {e}")

# Step 2: Count and print total number of lines
try:
    with open("python_notes.txt", "r", encoding="utf-8") as file:
        lines = file.readlines()
    
    total_lines = len(lines)
    print(f"\nTotal number of lines in the file: {total_lines}")
    
except FileNotFoundError:
    print("Error: File 'python_notes.txt' not found.")
except Exception as e:
    print(f" Error reading file: {e}")

# Step 3: Ask user for keyword and print matching lines (case-insensitive)
print("\n" + "-"*70)
keyword = input("Enter a keyword to search in the file: ").strip()

try:
    with open("python_notes.txt", "r", encoding="utf-8") as file:
        lines = file.readlines()
    
    matching_lines = []
    for line in lines:
        if keyword.lower() in line.lower():
            matching_lines.append(line.strip())
    
    if matching_lines:
        print(f"\nFound {len(matching_lines)} line(s) containing '{keyword}':\n")
        for line in matching_lines:
            print(f"  • {line}")
    else:
        print(f"\nℹ No lines found containing '{keyword}'. Try a different keyword.")
    
except FileNotFoundError:
    print(" Error: File 'python_notes.txt' not found.")
except Exception as e:
    print(f"Error reading file: {e}")
```

```python
# Task 2 - API Integration

import requests

print("TASK 2 - API INTEGRATION")
print("="*80 + "\n")

# Step 1 - Fetch and Display Products

print("=== STEP 1: FETCH AND DISPLAY PRODUCTS ===\n")

try:
    # Make GET request to fetch 20 products
    url = "https://dummyjson.com/products?limit=20"
    response = requests.get(url)
    response.raise_for_status()  # Raise exception for bad status codes
    
    # Parse JSON response
    data = response.json()
    products = data.get("products", [])
    
    print(f"Successfully fetched {len(products)} products\n")
    
    # Print formatted table
    print(f"{'ID':<5} | {'Title':<35} | {'Category':<15} | {'Price':<10} | {'Rating'}")
    print("-" * 80)
    
    for product in products:
        product_id = product.get("id", "N/A")
        title = product.get("title", "N/A")[:35]  # Truncate long titles
        category = product.get("category", "N/A")
        price = product.get("price", 0)
        rating = product.get("rating", 0)
        
        print(f"{product_id:<5} | {title:<35} | {category:<15} | ${price:<9.2f} | {rating}")
    
except requests.exceptions.RequestException as e:
    print(f" Error fetching products: {e}")
    products = []
except Exception as e:
    print(f"Unexpected error: {e}")
    products = []

# Step 2 - Filter and Sort

print("\n" + "="*80)
print("=== STEP 2: FILTER AND SORT ===\n")

try:
    # Filter products with rating >= 4.5
    filtered_products = [p for p in products if p.get("rating", 0) >= 4.5]
    
    print(f"Products with rating ≥ 4.5: {len(filtered_products)} found\n")
    
    # Sort by price in descending order
    sorted_products = sorted(filtered_products, key=lambda x: x.get("price", 0), reverse=True)
    
    # Print filtered and sorted list
    print(f"{'ID':<5} | {'Title':<35} | {'Category':<15} | {'Price':<10} | {'Rating'}")
    print("-" * 80)
    
    for product in sorted_products:
        product_id = product.get("id", "N/A")
        title = product.get("title", "N/A")[:35]
        category = product.get("category", "N/A")
        price = product.get("price", 0)
        rating = product.get("rating", 0)
        
        print(f"{product_id:<5} | {title:<35} | {category:<15} | ${price:<9.2f} | {rating}")

except Exception as e:
    print(f"Error filtering/sorting products: {e}")

# Step 3 - Search by Category

print("\n" + "="*80)
print("=== STEP 3: SEARCH BY CATEGORY (LAPTOPS) ===\n")

try:
    # Fetch products in 'laptops' category
    category_url = "https://dummyjson.com/products/category/laptops"
    response = requests.get(category_url)
    response.raise_for_status()
    
    # Parse JSON response
    data = response.json()
    laptops = data.get("products", [])
    
    print(f"Successfully fetched {len(laptops)} laptops\n")
    
    # Print name and price of each laptop
    print(f"{'Product Name':<40} | {'Price'}")
    print("-" * 55)
    
    for laptop in laptops:
        name = laptop.get("title", "N/A")
        price = laptop.get("price", 0)
        print(f"{name:<40} | ${price:.2f}")

except requests.exceptions.RequestException as e:
    print(f" Error fetching laptops: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")

# Step 4 - POST Request (Simulated)

print("\n" + "="*80)
print("=== STEP 4: POST REQUEST (SIMULATED) ===\n")

try:
    # POST request data
    post_url = "https://dummyjson.com/products/add"
    new_product = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }
    
    # Send POST request
    response = requests.post(post_url, json=new_product)
    response.raise_for_status()
    
    # Parse response
    created_product = response.json()
    
    print("POST request successful!\n")
    print("Full response from server:")
    print("-" * 80)
    
    # Print the full response
    for key, value in created_product.items():
        print(f"{key}: {value}")
    
    print("\nNote: DummyJSON is a mock/test API. The POST endpoint simulates a create")
    print("   operation and returns a response with an assigned ID, but no data is")
    print("   actually stored on the server. This is standard behaviour for test APIs.")

except requests.exceptions.RequestException as e:
    print(f"Error making POST request: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")
```

```python
# Task 3 - Exception Handling

import requests

print("TASK 3 - EXCEPTION HANDLING")
print("="*80 + "\n")

# Part A - Guarded Calculator

print("=== PART A: GUARDED CALCULATOR ===\n")

def safe_divide(a, b):
    """
    Safely divides a by b with error handling.
    Returns result or error message string.
    """
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

# Test the function
print("Testing safe_divide function:\n")
print(f"safe_divide(10, 2) = {safe_divide(10, 2)}")
print(f"safe_divide(10, 0) = {safe_divide(10, 0)}")
print(f"safe_divide('ten', 2) = {safe_divide('ten', 2)}")

# Part B - Guarded File Reader

print("\n" + "="*80)
print("=== PART B: GUARDED FILE READER ===\n")

def read_file_safe(filename):
    """
    Safely reads a file and returns its content.
    Handles FileNotFoundError and uses finally block.
    """
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    finally:
        print("File operation attempt complete.")

# Test with existing file
print("Test 1: Reading 'python_notes.txt' (should succeed):\n")
content1 = read_file_safe("python_notes.txt")
if content1:
    print(f"\nFile content retrieved ({len(content1)} characters)\n")

# Test with non-existing file
print("\nTest 2: Reading 'ghost_file.txt' (should fail gracefully):\n")
content2 = read_file_safe("ghost_file.txt")
if not content2:
    print("\nFile not found (handled gracefully)\n")

# Part C - Robust API Calls

print("="*80)
print("=== PART C: ROBUST API CALLS ===\n")

print("Making robust API calls with exception handling:\n")

# Example 1: Valid GET request
print("Test 1: Valid GET request to fetch products\n")
try:
    response = requests.get("https://dummyjson.com/products?limit=5", timeout=5)
    response.raise_for_status()
    data = response.json()
    print(f" Successfully fetched {len(data.get('products', []))} products\n")
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.\n")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.\n")
except Exception as e:
    print(f"Unexpected error: {e}\n")

# Example 2: Simulating timeout (with very short timeout)
print("Test 2: Request with very short timeout (simulating timeout)\n")
try:
    response = requests.get("https://dummyjson.com/products?limit=20", timeout=0.001)
    response.raise_for_status()
    print("Request succeeded\n")
except requests.exceptions.ConnectionError:
    print(" Connection failed. Please check your internet.\n")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.\n")
except Exception as e:
    print(f"Unexpected error: {e}\n")

# Example 3: Valid POST request
print("Test 3: Valid POST request\n")
try:
    new_product = {
        "title": "Test Product",
        "price": 499,
        "category": "electronics"
    }
    response = requests.post("https://dummyjson.com/products/add", json=new_product, timeout=5)
    response.raise_for_status()
    print(" POST request succeeded\n")
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.\n")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.\n")
except Exception as e:
    print(f" Unexpected error: {e}\n")

# Part D - Input Validation Loop

print("="*80)
print("=== PART D: INPUT VALIDATION LOOP ===\n")

print("Product Lookup Tool")
print("-" * 40)

while True:
    # Ask user for product ID
    user_input = input("\nEnter a product ID to look up (1-100), or 'quit' to exit: ").strip()
    
    # Check if user wants to quit
    if user_input.lower() == 'quit':
        print("\nExiting product lookup. Goodbye!")
        break
    
    # Validate input is an integer
    try:
        product_id = int(user_input)
    except ValueError:
        print(" Warning: Please enter a valid integer between 1-100.")
        continue
    
    # Validate input is in range 1-100
    if not (1 <= product_id <= 100):
        print("Warning: Product ID must be between 1 and 100. Please try again.")
        continue
    
    # Make API call to fetch product
    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)
        
        # Check status code
        if response.status_code == 404:
            print(f"Product not found (ID: {product_id})")
        elif response.status_code == 200:
            product = response.json()
            title = product.get('title', 'N/A')
            price = product.get('price', 0)
            print(f"\nProduct found!")
            print(f"   Title: {title}")
            print(f"   Price: ${price:.2f}")
        else:
            print(f" Unexpected response code: {response.status_code}")
    
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print(" Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
```

```python
# Task 4 - Logging to File

import requests
from datetime import datetime


print("TASK 4 - LOGGING TO FILE")
print("\n")

def log_error(function_name, error_message):
    """
    Logs an error to error_log.txt with timestamp.
    Format: [YYYY-MM-DD HH:MM:SS] ERROR in function_name: error_message
    """
    try:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log_entry = f"[{timestamp}] ERROR in {function_name}: {error_message}\n"
        
        # Open file in append mode so logs accumulate
        with open("error_log.txt", "a", encoding="utf-8") as log_file:
            log_file.write(log_entry)
        
        print(f"Logged error to error_log.txt")
    
    except Exception as e:
        print(f" Failed to write to log file: {e}")

# Test 1: Trigger ConnectionError by requesting unreachable URL

print("Test 1: Triggering ConnectionError with unreachable URL\n")

try:
    response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
    response.raise_for_status()
    print(" Request succeeded")

except requests.exceptions.ConnectionError as e:
    error_msg = "ConnectionError - No connection could be made"
    print(f"{error_msg}")
    log_error("fetch_products", error_msg)

except requests.exceptions.Timeout as e:
    error_msg = "Timeout - Request timed out"
    print(f" {error_msg}")
    log_error("fetch_products", error_msg)

except Exception as e:
    error_msg = f"Unexpected error - {str(e)}"
    print(f" {error_msg}")
    log_error("fetch_products", error_msg)

# Test 2: Trigger HTTP 404 error by requesting non-existent product

print("\n")
print("\nTest 2: Triggering HTTP 404 error with non-existent product ID\n")

try:
    product_id = 999
    response = requests.get(f"https://dummyjson.com/products/{product_id}", timeout=5)
    
    # Check for HTTP errors (404, etc.)
    if response.status_code != 200:
        error_msg = f"HTTPError - {response.status_code} Not Found for product ID {product_id}"
        print(f"{error_msg}")
        log_error("lookup_product", error_msg)
    else:
        print(" Product found")

except requests.exceptions.ConnectionError as e:
    error_msg = "ConnectionError - No connection could be made"
    print(f" {error_msg}")
    log_error("lookup_product", error_msg)

except requests.exceptions.Timeout as e:
    error_msg = "Timeout - Request timed out"
    print(f" {error_msg}")
    log_error("lookup_product", error_msg)

except Exception as e:
    error_msg = f"Unexpected error - {str(e)}"
    print(f" {error_msg}")
    log_error("lookup_product", error_msg)

# Read and print the full contents of error_log.txt


print("CONTENTS OF error_log.txt")
print("\n")

try:
    with open("error_log.txt", "r", encoding="utf-8") as log_file:
        contents = log_file.read()
    
    if contents:
        print(contents)
    else:
        print("(Log file is empty)")

except FileNotFoundError:
    print("error_log.txt not found (no errors have been logged yet)")

except Exception as e:
    print(f" Error reading log file: {e}")
```
